## Tendência do meio de pagamento

### Quais os meios de pagamento mais utilizados?

Meios de pagamento mais de forma geral

```python
plt.figure(dpi=250)
ax = sns.barplot(data=df_order_payments['payment_type'].value_counts(),
                 palette=palette, edgecolor=palette[4])

ax.set_ylabel('Quantidade')
ax.set_xlabel('Tipo de pagamento')
ax.set_title('Distribuição dos meios de pagamento por quantidade de transações')

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

Meios de pagamento com maiores gastos

```python
plt.figure(dpi=250)
ax = sns.barplot(
    df_order_payments.groupby(["payment_type"])["payment_value"]
    .sum()
    .sort_values(ascending=False),
    palette=palette,
    edgecolor=palette[4],
)

ax.set_ylabel("Valor (R$)")
ax.set_xlabel("Tipo de pagamento")
ax.set_title("Distribuição dos meios de pagamento por valor da transação")
ax.ticklabel_format(style="plain", axis="y")

plt.show()
```

Seguindo nas visualizações acima, criamos um gráfico que mostra a distribuição percentual de cada um destes itens:

```python
df_agg = (
    df_order_payments.groupby(["payment_type"])["payment_value"]
    .aggregate(["sum", "count"])
    .sort_values(by="count", ascending=False)
)

df_agg["valor"] = round(df_agg["sum"] / df_agg["sum"].sum() * 100, 3)
df_agg["quantidade"] = round(df_agg["count"] / df_agg["count"].sum() * 100, 3)

df_agg.reset_index(inplace=True)

df_melted = df_agg.melt(
    id_vars="payment_type",
    value_vars=["quantidade", "valor"],
    var_name="Métrica",
    value_name="Porcentagem",
)

plt.figure(figsize=(10, 6), dpi=250)
ax = sns.barplot(
    x="payment_type",
    y="Porcentagem",
    hue="Métrica",
    data=df_melted,
    palette=sns.color_palette("YlGnBu_d", n_colors=df_melted["payment_type"].nunique()),
    edgecolor=palette[4],
)

ax.set_ylabel("Porcentagem (%)")
ax.set_xlabel("Tipo de pagamento")
ax.set_title("Distribuição percentual dos meios de pagamento por quantidade e valor")

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

Observa-se que o percentual obtido em ambos os casos são próximos, indicando uma relação direta entre quantidade de usos e o valor total das compras.

Além disso, tanto via quantidade quanto pelo valor dos pedidos, o cartão de crédito fica bem na frente do que outros meios de pagamento, representando mais de 70% dos dados.

Vimos que o cartão de crédito é amplamente utilizado, estando presente na maioria das compras e também sendo responsável pela maior quantidade de dinheiro gasto.

Isso significaria que o cartão é o principal meio de pagamento em todas as faixas de preços? Ou ele seria apenas utilizado para produtos mais caros?

### Quais os meios de pagamento mais utilizados por faixa de preços?

Primeiro, vamos criar as faixas de preços:

```python
df_order_payments['payment_value'].describe()
```

```python
df_order_payments.query("payment_value == 0")
```

```python
sns.histplot(df_order_payments['payment_value'], bins=50, kde=True)
plt.title('Distribuição de payment_value')
plt.xlabel('Valor de pagamento')
plt.ylabel('Frequência')
plt.show()
```

```python
bins = [0, 50, 100, 200, 500, 1000, df_order_payments['payment_value'].max()]
labels = ['0-50', '50-100', '100-200', '200-500', '500-1000', '1000+']

df_order_payments['faixa_pagamento'] = pd.cut(df_order_payments['payment_value'], bins=bins, labels=labels, include_lowest=True)
df_order_payments['faixa_pagamento'].value_counts()
```

```python
plt.figure(figsize=(12, 6), dpi=250)
ax = sns.barplot(
    df_order_payments["faixa_pagamento"].value_counts(),
    palette=palette,
    edgecolor=palette[4],
)

ax.set_ylabel("Quantidade de transações")
ax.set_xlabel("Faixa de valor de pagamento (R$)")
ax.set_title("Distribuição de transações por faixa de valor")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=250)
ax = sns.countplot(
    x="faixa_pagamento", hue="payment_type", data=df_order_payments, palette=sns.color_palette("YlGnBu_d", n_colors=df_order_payments["payment_type"].nunique())
)

ax.set_ylabel("Quantidade de transações")
ax.set_xlabel("Faixa de valor de pagamento (R$)")
ax.set_title("Distribuição de transações por tipo de pagamento e faixa de valor")

plt.xticks(rotation=45)
plt.show()
```

```python
df_grouped = df_order_payments.groupby(['faixa_pagamento', 'payment_type']).size().reset_index(name='count')
df_grouped['percent'] = df_grouped.groupby('faixa_pagamento')['count'].transform(lambda x: x / x.sum() * 100)
df_grouped.sort_values(by=['faixa_pagamento', 'percent'], ascending=[True, False], inplace=True)
df_grouped #.query('faixa_pagamento == "1000+"')
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(
    x="faixa_pagamento",
    y="percent",
    hue="payment_type",
    data=df_grouped,
    palette=sns.color_palette(
        "YlGnBu_d", n_colors=df_grouped["payment_type"].nunique()
    ),
)

ax.set_ylabel("Percentual de transações")
ax.set_xlabel("Faixa de valor de pagamento (R$)")
ax.set_title("Distribuição de transações por tipo de pagamento e faixa de valor")

plt.xticks(rotation=45)
plt.show()
```

```python
df_grouped["percent_cum"] = df_grouped.groupby("faixa_pagamento")["percent"].cumsum()

plt.figure(figsize=(12, 8), dpi=250)

for payment_type, pal in zip(
    df_grouped["payment_type"].unique(),
    sns.color_palette("YlGnBu_d", n_colors=df_melted["payment_type"].nunique()),
):
    subset = df_grouped[df_grouped["payment_type"] == payment_type]
    plt.bar(
        subset["faixa_pagamento"],
        subset["percent"],
        bottom=subset["percent_cum"] - subset["percent"],
        label=payment_type,
        color=pal,
    )

for payment_type in df_grouped["payment_type"].unique():
    subset = df_grouped[df_grouped["payment_type"] == payment_type]
    for i in range(len(subset)):
        percent_value = subset["percent"].iloc[i]
        if percent_value >= 2:
            plt.text(
                subset["faixa_pagamento"].iloc[i],
                subset["percent_cum"].iloc[i]
                - subset["percent"].iloc[i] / 2,  # Centralizar na barra
                f"{percent_value:.2f}%",
                ha="center",
                va="center",
                color="white",
            )

plt.ylabel("Percentual de transações")
plt.xlabel("Faixa de valor de pagamento (R$)")
plt.title(
    "Distribuição de transações por tipo de pagamento e faixa de valor (100% Empilhado)"
)
plt.xticks(rotation=45)
plt.legend(title="Tipo de pagamento")
plt.tight_layout()

plt.show()
```

Algumas conclusões sobre a análise dos meios de pagamento:

- Sobre o cartão de crédito:
    - Ele domina todas as faixas, mas aparece menos em transações de valores mais baixos
    - Isso indicaria que as pessoas preferem utilizar menos o crédito em valores menores, dando preferência a outros meios
- Sobre o boleto:
    - Possui uma proporção próxima em todas as faixas
    - Aparece menos para valores mais altos, talvez indicando uma certa dificuldade do brasileiro conseguir pagar compras mais caras no boleto
- Sobre o voucher:
    - Aparece bem nas faixas mais baixas, mas diminui bastante a partir de R$ 100,00.
    - Talvez isso indique uma dificuldade em acessar vouchers para compras mais caras, pois geralmente vouchers seguem regras
- Sobre o debit_card:
    - Aparece pouco na base, nunca representando mais do que 2% de uma faixa
    - Similar ao boleto, aparece mais nas faixas mais baixas e diminui nas mais altas
    - Nas transações acima de R$ 1000,00, o cartão de débito aparece pela primeira vez maior do que o voucher

## Tendência sobre os pedidos

### Como é a distribuição de pedidos por data?

Aqui é interessante analisar se há um aumento de vendas/pedidos em certos períodos.

Para a nossa análise é importante filtrar os pedidos, evitando aqueles que foram cancelados ou estão indisponíveis.

```python
plt.figure(dpi=250)
ax = sns.barplot(data=df_orders['order_status'].value_counts(),
                 palette=palette, edgecolor=palette[4])

ax.set_ylabel('Quantidade')
ax.set_xlabel('Status')
ax.set_title('Distribuição dos status por pedido')

for i in ax.containers:
    ax.bar_label(i)

plt.xticks(rotation=45)
plt.show()
```

```python
df_orders = df_orders.query("order_status not in ['canceled', 'unavailable']")
df_orders
```

```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['day_order_purchase_timestamp'] = df_orders['order_purchase_timestamp'].dt.normalize()
df_day = df_orders.groupby(['day_order_purchase_timestamp']).size().reset_index(name='count')
df_day['moving_avg'] = df_day['count'].rolling(window=7).mean()
df_day.sort_values(by='count', ascending=False).head(10)
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(
    x="day_order_purchase_timestamp",
    y="count",
    data=df_day,
    label="Contagem de Pedidos",
    marker="",
    color=palette[3],
)
sns.lineplot(x='day_order_purchase_timestamp', y='moving_avg', data=df_day, color=palette[0],
             label='Média Móvel (7 dias)', linewidth=2, marker='.')

plt.xlabel("Data")
plt.ylabel("Contagem de Pedidos")
plt.title("Tendência de Pedidos ao Longo do Tempo (Diário)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

> Houve um pico nas compras em e-commerces brasileiros no dia 24/11/2017 (Black Friday)

```python
df_orders['month'] = df_orders['order_purchase_timestamp'].dt.to_period('M')

df_monthly = df_orders.groupby('month').agg({'order_purchase_timestamp': 'first', 'order_id': 'size'}).reset_index()
df_monthly['order_purchase_timestamp'] = pd.to_datetime(df_monthly['order_purchase_timestamp']).dt.normalize()
df_monthly.rename(columns={'order_id': 'count'}, inplace=True)
df_monthly.sample(5)
```

```python
df_monthly['moving_avg'] = df_monthly['count'].rolling(window=3).mean()

plt.figure(figsize=(12, 8), dpi=250)
sns.lineplot(x='order_purchase_timestamp', y='count', data=df_monthly, label='Contagem de Pedidos', marker='o', color=palette[3])
sns.lineplot(x='order_purchase_timestamp', y='moving_avg', data=df_monthly, color=palette[0],
             label='Média Móvel (3 meses)', linewidth=2, marker='.')

for x, y in zip(df_monthly['order_purchase_timestamp'], df_monthly['count']):
    plt.text(x, y+50, f'{int(y)}', ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Data')
plt.ylabel('Contagem de Pedidos')
plt.title('Pedidos ao Longo do Tempo (Mensal)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
num_months = 6
df_monthly['month_change'] = df_monthly['count'].pct_change().fillna(0)
print('Meses com uma alta diferença quando comparados ao mês passado (com outliers):')
df_monthly.nlargest(n=num_months, columns='month_change')
```

Aplicando remoção de outliers com IQR para identificar os meses com uma maior variação em relação ao seu mês passado:

```python
Q1 = df_monthly["month_change"].quantile(0.25)
Q3 = df_monthly["month_change"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print('Meses com uma alta diferença quando comparados ao mês passado (sem outliers):')

highest_month_changes = df_monthly[
    (df_monthly["month_change"] >= lower_bound)
    & (df_monthly["month_change"] <= upper_bound)
].nlargest(n=num_months, columns="month_change")
highest_month_changes
```

```python
plt.figure(figsize=(12, 8), dpi=250)

sns.lineplot(
    x="order_purchase_timestamp",
    y="count",
    data=df_monthly,
    label="Contagem de Pedidos",
    marker="o",
    color=palette[3],
)
sns.lineplot(
    x="order_purchase_timestamp",
    y="moving_avg",
    data=df_monthly,
    color=palette[0],
    label="Média Móvel (3 meses)",
    linewidth=2,
    marker=".",
)
plt.scatter(
    highest_month_changes["order_purchase_timestamp"],
    highest_month_changes["count"],
    color="red",
    s=30,
    zorder=5,
    label="Meses com alto crescimento",
    edgecolor="black",
)

for x, y in zip(df_monthly["order_purchase_timestamp"], df_monthly["count"]):
    plt.text(
        x, y + 50, f"{int(y)}", ha="center", va="bottom", fontsize=10, color="black"
    )

plt.xlabel("Data")
plt.ylabel("Contagem de Pedidos")
plt.title("Pedidos ao Longo do Tempo (Mensal)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Analisando os valores dos gráficos e tabelas acima, aparentemente o dataset não possui muitos valores de 2016 e do final de 2018, tornando-se períodos não muito confiáveis para analisar. De qualquer forma, pode-se observar algumas coisas:

- A média móvel conseguiu seguir de forma boa o ritmo dos meses, indicando que a quantidade de pedidos prevista em 2018 provavelmente foi maior
- Com exceção do final de 2018 (provavelmente dados faltantes), a média do número de pedidos foi ficando mais alta conforme o tempo, indicando, talvez, o sucesso e popularidade do e-commerce brasileiro que vemos nos anos atuais
- Em 2017, temos alguns momentos que chamam atenção, onde os valores aumentam bastante:
    - Em maio (3640 pedidos), comparado ao valor anterior de abril (2377 pedidos)
    - Em novembro (7423), comparado a outubro (4527)
    - Em janeiro (7187), comparado a dezembro (5620)
- Esses picos de pedidos podem ter sido causados por eventos sazonais, como:
    - Maio:
        - Dia das mães
    - Novembro - mês com maior pico de vendas:
        - Black Friday
        - Cyber Monday (primeira segunda após a Black Friday)
    - Janeiro:
        - Liquidações de verão
        - Volta às aulas

Com isso em mente, seria bom entender melhor como as vendas se comportam nesse período, verificando se há uma tendência no consumo dos brasileiros nestes no geral, mas também nestes períodos de pico.

### Quais as tendências de produtos nos pedidos?

Primeiro, precisamos juntar tudo em uma só base, fazendo um merge entre:
- Pedidos (`df_orders`) - para buscar as informações gerais do pedido
- Itens de cada pedido (`df_order_items`) - para buscar quais produtos estão no pedido
- Produtos (`df_products`) - para obter as categorias dos produtos do pedido

```python
orders_products = df_orders.merge(df_order_items, on='order_id', how='left').\
    merge(df_products[['product_id', 'product_category_name']], on='product_id', how='left')

orders_products
```

#### Como foi a distribuição dos produtos geral e nos meses com picos de venda?

> Não faz muito sentido ter uma análise por **dia** sobre as categorias dos produtos, pois isso teria o objetivo de identificar se em algum dia houve uma maior venda de algum produto específico e, para a nossa quantia de dados, uma análise por mês se mostra mais eficiente

```python
orders_products['order_purchase_timestamp'] = pd.to_datetime(orders_products['order_purchase_timestamp'])
orders_products['month'] = orders_products['order_purchase_timestamp'].dt.to_period('M')
orders_products[['order_purchase_timestamp', 'month']].sort_values(by='month')
```

```python
plt.figure(figsize=(14, 6), dpi=250)

ax = sns.barplot(
    orders_products["product_category_name"].value_counts().nlargest(15).reset_index(),
    x="product_category_name",
    y="count",
    color=palette[1],
    edgecolor=palette[4],
)

ax.set_ylabel("Número de produtos")
ax.set_xlabel("Categoria do produto")
ax.set_title("Categorias que mais aparecem nos pedidos")

plt.xticks(rotation=90)
plt.show()
```
